clustering

```python
true_labels =df['Class']
```

```python
df.drop(['Class'], axis=1)
```

```python
scaler = StandardScaler()
df_normalized = scaler.fit_transform(df)
```

```python
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_normalized)
Kmeans = KMeans(n_clusters=2)
kmeans = Kmeans.fit(df_pca)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
colors = ["g.", "r.", "c.", "b.", "y."]
for i in range(len(df_pca)):
    plt.plot(df_pca[i][0], df_pca[i][1], colors[labels[i]], markersize=10)

plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=150, linewidths=5, zorder=10)
plt.title("K-Means Clustering")
plt.show()
```

```python
kmeans1 = KMeans(n_clusters=2).fit_predict(df)
```

```python
conf_matrix = confusion_matrix(true_labels, kmeans1)

# Display the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix for K-Means Clustering')
plt.show()
```

```python
df['Cluster'] = kmeans.labels_


plt.figure(figsize=(10, 6))
parallel_coordinates(df, 'Cluster', colormap='viridis')
plt.title('K-Means Clustering - Parallel Coordinates')
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score , accuracy_score
```

```python
# Accuracy
accuracy = accuracy_score(true_labels, kmeans1)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.datasets import make_blobs
df, true_labels = make_blobs(centers=31, random_state=42)

# Apply k-means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
predicted_labels = kmeans.fit_predict(df)

# Adjusted Rand Index (ARI)
ari = adjusted_rand_score(true_labels, predicted_labels)
print(f"Adjusted Rand Index (ARI): {ari}")

# Normalized Mutual Information (NMI)
nmi = normalized_mutual_info_score(true_labels, predicted_labels)
print(f"Normalized Mutual Information (NMI): {nmi}")
```

```python
df = pd.read_csv('/content/creditcard.csv')
```

```python
class_counts = df['Class'].value_counts()
fraud_percentage = (class_counts[1] / len(df)) * 100
non_fraud_percentage = (class_counts[0] / len(df)) * 100
print(f"Number of Fraud data: {class_counts[1]}")
print(f"Number of Non-Fraud data: {class_counts[0]}")
print(f"Percentage of Fraud data: {fraud_percentage}")
print(f"Percentage of Non-Fraud data: {non_fraud_percentage}")
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd


y = df['Class']
X = df.drop('Class', axis=1)


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Evaluate the performance of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Display classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.datasets import make_classification
import numpy as np


desired_percentage_class_0 = 85
desired_percentage_class_1 = 15
total_sample=len(df)
num_class_0 = int(total_sample * desired_percentage_class_0 / 100)
num_class_1 = total_sample - num_class_0


X, y = make_classification(
    n_samples=total_sample,
    weights=[desired_percentage_class_0 / 100, desired_percentage_class_1 / 100],

)
df = pd.DataFrame(X, columns=[f'feature_{i}' for i in range(X.shape[1])])
df['Class'] = y
# X = df.drop('Class', axis=1)
# Display the actual class distribution in the synthetic dataset
# Display the actual class distribution in the synthetic dataset
actual_percentage_class_0 = np.sum(df['Class'] == 0) /total_sample * 100
actual_percentage_class_1 = np.sum(df['Class'] == 1) / total_sample * 100

print(f"Actual Percentage for Class 0: {actual_percentage_class_0:.2f}%")
print(f"Actual Percentage for Class 1: {actual_percentage_class_1:.2f}%")
```

```python
class_counts = df['Class'].value_counts()
fraud_percentage = (class_counts[1] / len(df)) * 100
non_fraud_percentage = (class_counts[0] / len(df)) * 100
print(f"Number of Fraud data: {class_counts[1]}")
print(f"Number of Non-Fraud data: {class_counts[0]}")
print(f"Percentage of Fraud data: {fraud_percentage}")
print(f"Percentage of Non-Fraud data: {non_fraud_percentage}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class', axis=1), df['Class'], test_size=0.2, random_state=42)


model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

# Predict on the testing set
y_pred = model.predict(X_test)

# Evaluate the performance of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Display classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class', axis=1), df['Class'], test_size=0.2, random_state=42)


rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

# Evaluate the performance of the Random Forest model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf:.2f}")

# Display classification report for Random Forest
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
```
